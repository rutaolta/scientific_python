### Running FVA

```python
# FBA will not always give a unique solution, because multiple flux states can achieve the same optimum.
# FVA (or flux variability analysis) finds the ranges of each metabolic flux at the optimum.
```

```python
from cobra.flux_analysis import flux_variability_analysis
```

```python
flux_variability_analysis(model, model.reactions[:10])
```

```python
# Setting parameter fraction_of_optimium=0.90 would give the flux ranges for reactions at 90% optimality.

flux_variability_analysis(model, model.reactions[:10], fraction_of_optimum=0.9)
```

```python
# Flux variability analysis can also be embedded in calls to summary methods. 
# For instance, the expected variability in substrate consumption and product formation can be quickly found by

model.optimize()
model.summary(fva=0.95)
```

```python
# Similarly, variability in metabolite mass balances can also be checked with flux variability analysis:

model.metabolites.atp_m.summary(fva=0.95)
```

```python
# Parsimonious FBA (often written pFBA) finds a flux distribution which gives the optimal growth rate, 
# but minimizes the total sum of flux:

cobra.flux_analysis.pfba(model)
```

```python
# In most cases it doesn't change flux distribution a lot:
model.objective = 'biomass'
fba_solution = model.optimize()
pfba_solution = cobra.flux_analysis.pfba(model)
```

```python
abs(fba_solution.fluxes["biomass"] - pfba_solution.fluxes["biomass"])
```

```python
'''
There are three different types of pre-defined boundary reactions: exchange, demand, and sink reactions.
An exchange reaction is a reversible, imbalanced reaction that adds to or removes an extracellular metabolite 
from the extracellular compartment. A demand reaction is an irreversible reaction that consumes an
intracellular metabolite. A sink is similar to an exchange but specifically for intracellular metabolites.
https://cobrapy.readthedocs.io/_/downloads/en/0.10.0/pdf/
'''

# model.exchanges
# model.demands
# model.sinks
# model.boundary
model.medium
```
